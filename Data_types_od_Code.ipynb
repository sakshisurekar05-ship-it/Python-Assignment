```python
#Explain Datatypes in Python

#Python Data Types are used to store different kinds of values.

#Main Datatypes in Python
#| Datatype | Example    | Description              |
#| int      | 10         | Integer numbers          |
#| float    | 10.5       | Decimal numbers          |
#| str      | "Hello"    | Text/String              |
#| bool     | True/False | Boolean values           |
#| list     | [1,2,3]    | Ordered and changeable   |
#| tuple    | (1,2,3)    | Ordered and unchangeable |
#| dict     | {"a":1}    | Key-value pairs          |
#| set      | {1,2,3}    | Unordered unique values  |

#Difference Between List, Tuple, Dictionary, NumPy
#| Feature        | List            | Tuple      | Dictionary   | NumPy Array          |
#| -------------- | --------------- | ---------- | ------------ | -------------------- |
#| Syntax         | []              | ()         | {}           | np.array()           |
#| Mutable        | Yes             | No         | Yes          | Yes                  |
#| Ordered        | Yes             | Yes        | Yes          | Yes                  |
#| Key-Value Pair | No              | No         | Yes          | No                   |
#| Faster         | Normal          | Faster     | Fast lookup  | Very Fast            |
#| Use            | General storage | Fixed data | Data mapping | Numerical operations |

import numpy as np
# list example
my_list = [1, 2, 3]
print(my_list)
#tuple example
my_tuple = (1, 2, 3)
print(my_tuple)
#dictionary example
my_dict = {"name": "Sakshi", "age": 19}
print(my_dict)
#numpy example
my_array = np.array([1, 2, 3])
print(my_array)
```
